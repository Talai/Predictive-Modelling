```python
N = 100
tmin = -5.
tmax = 5.
t = np.linspace(tmin, tmax, N)
```

```python
w0=1
N=100
t=np.linspace(-5,5,N)
y=np.zeros(N)
for i in range(N):
    s=0
    for k in range(1,3):
        s+=np.sin(k*t[i]*w0)
        y[i]=1+(2*s)
plt.stem(t,y,'orange', label='sin()')
plt.legend()
plt.show()
```
